```python
# 尝试爬取mooc课程的课件
# 链接:https://www.icourse163.org/learn/HIT-309001?tid=1470861446#/learn/content
# 导包
import requests
import re
import os
import time
from bs4 import BeautifulSoup
```

```python
# 爬取所有pdf文件
# 1.获取网页源代码
response = requests.get('https://www.icourse163.org/learn/HIT-309001?tid=1470861446#/learn/content')
html = response.text
# https://www.icourse163.org/learn/HIT-309001?tid=1470861446#/learn/content?type=detail&id=1255023581&cid=1286611847
# https://www.icourse163.org/learn/HIT-309001?tid=1470861446#/learn/content?type=detail&id=1255023582&cid=1286611854
# https://www.icourse163.org/learn/HIT-309001?tid=1470861446#/learn/content?type=detail&id=1255023583&cid=1286611857
# https://www.icourse163.org/learn/HIT-309001?tid=1470861446#/learn/content?type=detail&id=1255023585&cid=1286611862
# 搜索html里面是否有:https://www.icourse163.org/learn/HIT-309001?tid=1470861446#/learn/content?type=detail&id=1255023581&cid=1286611847
print(html)
```

```python
# 解析HTML
soup = BeautifulSoup(html, 'html.parser')

# 找到所有的链接
links = soup.find_all('a')
print(links)
# 遍历所有的链接
for link in links:
    # 如果链接的文本包含"文档"
    if '文档' in link.text:
        # 打印链接的href属性，即URL
        print(link['href'])
```
