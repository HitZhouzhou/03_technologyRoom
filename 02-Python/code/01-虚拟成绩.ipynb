```python
import pandas as pd
import numpy as np
# 随机生成两组数据,每组45个,代表45个学生的成绩,每个学生对应1,2,3...45号,第一组代表第一次考试成绩,第二组代表第二次考试成绩,要求如下:
# 第一次平均分:62
# 第二次平均分:70
# 每个学生两次波动不超过10分

# 生成数据
np.random.seed(0)
data1 = np.random.normal(62, 5, 45).round(0)
data2 = np.random.normal(72, 5, 45).round(0)
```

```python
print(data1)
print("data1.mean = ", data1.mean())
print(data2)
print("data2.mean = ", data2.mean())
```

```python
# 把data1,data2合并成一个DataFrame
df = pd.DataFrame({'第一次考试成绩': data1, '第二次考试成绩': data2})
# 添加一列,代表学生的编号
df['学生编号'] = range(1, 46)
```

```python
df
```

```python
# 第二个班级的数据
data3 = np.random.normal(64, 5, 45).round(0)
data4 = np.random.normal(79, 5, 45).round(0)
# 把data3,data4合并成一个DataFrame
df2 = pd.DataFrame({'第一次考试成绩': data3, '第二次考试成绩': data4})
# 添加一列,代表学生的编号
df2['学生编号'] = range(1, 46)
```

```python
df2
```

```python
print(data3.mean())
print(data4.mean())
```

```python
df.to_excel('第一班成绩.xlsx', index=False)
df2.to_excel('第二班成绩.xlsx', index=False)
```
